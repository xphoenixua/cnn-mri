# Вступ

Актуальність проблеми виявлення пухлин головного мозку за допомогою комп'ютерних нейронних мереж особливо важлива в сучасному світі медичних технологій. Пухлини головного мозку представляють серйозний ризик для здоров'я людини, а їх своєчасне виявлення і точна діагностика є ключовими для ефективного лікування. За даними дослідження [1], пухлини мозку займають одне з провідних місць серед причин смертності та інвалідності у всьому світі. Однак раннє виявлення та точна класифікація пухлин можуть значно підвищити шанси на успішне лікування та поліпшення якості життя пацієнтів.

Традиційно діагностика пухлин головного мозку здійснювалася за допомогою медичних зображень, отриманих методами МРТ (магнітно-резонансної томографії) [2]. Проте, аналіз таких зображень вимагає високої кваліфікації та значного часу від фахівців, і навіть досвідчені лікарі можуть зіткнутися з викликами, пов'язаними з суб'єктивною інтерпретацією даних. Це призводить до потреби у розробці більш ефективних, точних і автоматизованих систем для аналізу медичних зображень.

Саме тому я використаю згорткові нейронні мереж для аналізу МРТ-зображень пухлин головного мозку. Розвиток і впровадження таких систем має потенціал революціонізувати область медичної візуалізації та надати потужний інструмент для підвищення ефективності та якості медичного обслуговування.

# Завдання

Завдання мого дослідження полягає у розробці та валідації комп'ютерної системи, що використовує методи глибокого навчання для ефективної класифікації МРТ-зображень на ті, що містять пухлини головного мозку, та ті, що не містять.

Зображення, що містять пухлини, зазвичай характеризуються низкою візуальних ознак, таких як ненормальні ущільнення тканини, зміни в структурі мозку, аномальні області з високою або низькою інтенсивністю сигналу, та інші нерегулярності, які відрізняються від нормальної анатомії. Водночас зображення без пухлини зазвичай показують рівномірну та очікувану структуру мозкової тканини без видимих аномалій. Розрізнення цих особливостей може бути непростим завданням, особливо у випадках, коли пухлини невеликі або мають нетиповий вигляд [3].

Цільовий набір даних було обрано із платформи Kaggle - Brain MRI Images for Brain Tumor Detection [4]. Приклади таких зображень буде наведено нижче, у пірозділі Попередня обробка даних пункті Завантаження даних.

## Попередня обробка даних

### Завантаження даних

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns

import cv2
import os
import shutil
```

```python
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from sklearn.model_selection import train_test_split
from PIL import Image

# директорії до класів
yes_dir = '/kaggle/input/brain-mri-images-for-brain-tumor-detection/yes'
no_dir = '/kaggle/input/brain-mri-images-for-brain-tumor-detection/no'

# допоміжна функція для підвантаження зображень
def load_images(directory):
    images = []
    sizes = []
    for filename in os.listdir(directory):
        if any(filename.lower().endswith(ext) for ext in ['.jpg', '.png', '.jpeg']):
            img_path = os.path.join(directory, filename)
            img = load_img(img_path, color_mode='grayscale') # МРТ є монохромними
            img_array = img_to_array(img)
            images.append(img_array)
            # тут ми зберігаємо розміри зображень для подальшого аналізу
            with Image.open(img_path) as img:
                sizes.append(img.size)
    return images, sizes

yes_images, yes_sizes = load_images(yes_dir)
no_images, no_sizes = load_images(no_dir)
```

```python
def plot_images(images, labels, rows=2, cols=5, title='Images'):
    plt.figure(figsize=(15, 5))
    for i in range(rows * cols):
        plt.subplot(rows, cols, i + 1)
        img = images[i].squeeze() if images[i].shape[-1] == 1 else images[i]
        plt.imshow(img, cmap='gray')
        plt.axis('off')
    plt.suptitle(title)
    plt.tight_layout()
    plt.show()
```

```python
plot_images(yes_images[10:], ['Yes'] * len(yes_images), rows=1, cols=5, title='Images with Tumor')
plot_images(no_images[10:], ['No'] * len(no_images), rows=1, cols=5, title='Images without Tumor')
```

### Дослідження розподілу розмірів зображень

```python
all_sizes = yes_sizes + no_sizes 
widths, heights = zip(*all_sizes)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(widths, bins=20, color='skyblue')
plt.title('Розподіл ширини зображень')
plt.xlabel('Ширина')
plt.ylabel('N')

plt.subplot(1, 2, 2)
plt.hist(heights, bins=20, color='salmon')
plt.title('Розподіл висоти зображень')
plt.xlabel('Висота')
plt.ylabel('N')

plt.tight_layout()
plt.show()
```

Видно, що всі зображення є більшими за 200 пікселів в обидва напрямки. Тому буде сенс масштабувати їх до квадратного розміру 224х224. Бонуси такого формату в нашому випадку:
- є найближчим до можливого найменшого зображення МРТ скану (буде більш ресурсно ефективно тренувати на менших зображеннях)
- цей розмір є ступенем двійки 2^7 (підійде до більшості архітектур, так як згорткові та пулінґові шари зменшують розмірність вхідних зображень на ступінь двійки)
- потенційне використання на претренованих моделях на ImageNet датасеті, це якраз не буде суперечити даним, що знаходяться у ньому.

```python
# функція для масштабування зображень
def resize_images(image_arrays, size=(224, 224)):
    resized_images = []
    for img_array in image_arrays:
        if img_array.ndim == 3 and img_array.shape[-1] == 1:
            img_array = img_array.reshape(img_array.shape[0], img_array.shape[1])
        img = Image.fromarray(np.uint8(img_array))  # конвертація у формат PIL Image
        img_resized = img.resize(size, Image.LANCZOS)
        img_resized_array = img_to_array(img_resized)  # і назад у пайтонівський масив
        resized_images.append(img_resized_array)
    return np.array(resized_images)
```

```python
yes_images_resized = resize_images(yes_images)
no_images_resized = resize_images(no_images)
```

```python
print(f'Нові виміри: {yes_images_resized.shape, no_images_resized.shape}')
```

### Розділення та кодування датасету

```python
# генерація міток для зображеннь (1/0 - є/немає пухлини)
yes_labels = np.ones(yes_images_resized.shape[0])
no_labels = np.zeros(no_images_resized.shape[0])

x = np.concatenate((yes_images_resized, no_images_resized), axis=0)
y = np.concatenate((yes_labels, no_labels), axis=0)

# розділення на тренувальний і тестовий набори (80% і 20% відповідно)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1337)
```

### Дослідження збалансованості класів

```python
labels = ['З пухлиною', 'Без пухлини']
values = [y_train[y_train == 1].shape[0], y_train[y_train == 0].shape[0]]

plt.figure(figsize=(7, 3))
plt.bar(labels, values, color=['skyblue', 'salmon'])
plt.title('Кількість зображень за класами')
plt.xlabel('Клас')
plt.ylabel('N')
plt.show()
```

Бачимо, що один клас переважає інший. Треба виправляти цю ситуацію, щоб позбутися упередження на клас знімків із пухлиною. Це можна здійснити аугментацією знімків, що без пухлини мозку.

### Аугментація класу без пухлини

```python
# параметри аугментації
augmenter = ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.05,
        zoom_range=0.1,
        brightness_range=[0.5, 1.5],
        horizontal_flip=True,
        fill_mode='constant',
        cval=0,
        rescale=1./255
)

augmenter.fit(x_train)

# різниця між класами після розділення
num_to_augment = y_train[y_train == 1].shape[0] - y_train[y_train == 0].shape[0]

# генерую аугментовані зображення
augmented_images, augmented_labels = next(augmenter.flow(x_train[y_train == 0], y_train[y_train == 0], batch_size=num_to_augment))

x_train = np.concatenate((x_train, augmented_images), axis=0)
y_train = np.concatenate((y_train, augmented_labels), axis=0).reshape(-1, 1)

print(f"Нові розміри: {x_train.shape}, {y_train.shape}")
```

# Перші моделі. Підбір швидкості навчання та функції активації.

Спробуємо використати архітектуру, яку я використовував і вона гарно справилася на CIFAR-10 датасеті. Хоча ці дві задачі є діаметрально протилежним (зараз ми маємо монохромні зображення 224х224 МРТ-сканів, а тоді я мав 32х32 кольорові зображень тварин та транспорту), все одно можна спробувати перебрати деякі параметри регуляризації або інші гіперпараметри та подивитися на результати моделі.

## Імпортування необхідних бібліотек

```python
from keras.models import Sequential
from keras.regularizers import l2
from keras.layers import Dropout, BatchNormalization, Conv2D, MaxPooling2D, Flatten, Dense
from keras.optimizers import Nadam, RMSprop
from keras.losses import BinaryCrossentropy
from keras.layers import Activation
from keras import backend as K
from sklearn.metrics import confusion_matrix
```

## Розробка моделі

```python
config_1 = {'name': 'Optimizers and Activations',
            'method': 'grid',
               'metric': {
                   'name': 'val_accuracy',
                   'goal': 'maximize'
                    },
                'parameters': {
                    'activation' : {
                        'values': ['leaky_relu', 'tanh']
                        },
                    'optimizer' : {
                        'values': ['Nadam', 'RMSprop']
                        },
                    'learning_rate': {
                        'values': [1e-2, 1e-3]
                        }
                    }
                }
```

```python
def f1_score(y_true, y_pred):

    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    
    f1_val = 2 * (precision * recall) / (precision + recall + K.epsilon())
    return f1_val
```

```python
def build_optimizer(opt, lr):
    parameter_mapper = {'Nadam': Nadam(learning_rate=lr),
                        'RMSprop': RMSprop(learning_rate=lr)}
    return parameter_mapper[opt]
```

```python
def build_activation(act):
    parameter_mapper = {'leaky_relu': tf.nn.leaky_relu,
                        'tanh': tf.nn.tanh}
    return parameter_mapper[act]
```

```python
def build_model(optimizer, activation, kernel_regularizer=None, dropout_rate=0.0):
    model = Sequential()
    
    # First Convolutional Block
    model.add(Conv2D(32, (3, 3), 
                     padding='same', 
                     kernel_regularizer=kernel_regularizer, 
                     input_shape=(224, 224, 1)))
    model.add(BatchNormalization())
    model.add(Activation(activation))
    model.add(Conv2D(32, (3, 3), 
                     padding='same', 
                     kernel_regularizer=kernel_regularizer))
    model.add(BatchNormalization())
    model.add(Activation(activation))
    model.add(MaxPooling2D((2, 2)))

    # Second Convolutional Block
    model.add(Conv2D(64, (3, 3), 
                     padding='same', 
                     kernel_regularizer=kernel_regularizer))
    model.add(BatchNormalization())
    model.add(Activation(activation))
    model.add(Conv2D(64, (3, 3), 
                     padding='same', 
                     kernel_regularizer=kernel_regularizer))
    model.add(BatchNormalization())
    model.add(Activation(activation))
    model.add(MaxPooling2D((2, 2)))

    # Third Convolutional Block (No MaxPooling)
    model.add(Conv2D(128, (3, 3), 
                     padding='same', 
                     kernel_regularizer=kernel_regularizer))
    model.add(BatchNormalization())
    model.add(Activation(activation))
    model.add(Conv2D(128, (3, 3), 
                     padding='same',
                     kernel_regularizer=kernel_regularizer))
    model.add(BatchNormalization())
    model.add(Activation(activation))

    # Flattening the output for the dense layers
    model.add(Flatten())

    # Dense Layers
    model.add(Dense(256, 
                    kernel_regularizer=kernel_regularizer))
    model.add(BatchNormalization())
    model.add(Activation(activation))
    model.add(Dropout(dropout_rate))
    model.add(Dense(128, 
                    kernel_regularizer=kernel_regularizer))
    model.add(BatchNormalization())
    model.add(Activation(activation))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))
    model.add(Activation(tf.nn.sigmoid))
                  
    model.compile(optimizer=optimizer,
                  loss=BinaryCrossentropy(),
                  metrics=['accuracy', f1_score])
    return model
```

## Тренування

Як можна бачити нижче, я використовую сервіс Weights&Biases, який допомагає відслідковувати різні комбінації параметрів моделі послідовно (хоча є і інші методи, як Random чи Bayes пошуки параметрів) тренуючи моделі на них.

```python
# !pip install wandb -q
import wandb
wandb.login()
```

```python
sweep_id = wandb.sweep(config_1, project="CNN-MRI")

def train(config=None):
    with wandb.init(config=config):
        config = wandb.config
        act = config.activation
        opt = config.optimizer
        lr = config.learning_rate
        wandb.run.name = f'{opt}({lr}) | {act}'
        optimizer = build_optimizer(opt, lr)
        activation = build_activation(act)
        
        keras.utils.set_random_seed(1337)
        
        model = build_model(optimizer=optimizer, activation=activation)
        model.fit(x_train, y_train, 
                  epochs=10, batch_size=64, 
                  validation_data=(x_test, y_test), 
                  verbose=0, callbacks=[wandb.keras.WandbCallback()])

wandb.agent(sweep_id=sweep_id, function=train)
wandb.finish()
```

### Аналіз результатів

Як можна спостерігати, швидкість навчання 1е-2 хоч і дала гарне тренування у Nadam, але загалом це занадто високе значення згідно графіку втрат для практично усіх моделей із такою швидкістю навчання, таке враження, ніби модель проскакує локальний мінімум, тим самим показує гіршу збіжність на відміну від моделей із нижчою швидкістю навчання.

Дуже важливим аспектом у цих результатах є перенавчання. Воно дуже ясно видно, дивлячись на перші кілька моделей: тренувальні метрики рівні 1, а валідаційні рівні достатньо нижчому значенню (хоча все одно не зовсім поганому). Тому це можна виправити шляхом  додавання регуляризації. У наступному блоці я буду тестувати L2-регуляризацію, а також значення частки викидання нейронів.

Щодо функції активації, то LeakyReLU показала себе краще за TanH.

Підсумовуючи, давайте оберемо Nadam(1e-3) із LeakyReLU для подальшого дослідження.

# Оптимізація отриманої моделі

## Підбір коефіцієнта регуляризації та частки виключених нейронів.

```python
config_2 = {'name': 'Regularization Control',
            'method': 'grid',
            'metric': {
                'name': 'val_accuracy',
                'goal': 'maximize'
                },
            'parameters': {
                'dropout': {
                    'values': [0.2, 0.5]
                    },
                'regularizer_coef': {
                    'values': [0, 1e-2]
                    }
                }
            }
```

```python
sweep_id = wandb.sweep(config_2, project="CNN-MRI")

def train(config=None):
    with wandb.init(config=config):
        config = wandb.config
        regularizer_coef = config.regularizer_coef
        dropout = config.dropout
        wandb.run.name = f'L2({regularizer_coef}, Dropout({dropout})'
        
        keras.utils.set_random_seed(1337)
        
        model = build_model(optimizer=Nadam(1e-3), activation=tf.nn.leaky_relu, 
                            kernel_regularizer=l2(regularizer_coef), dropout_rate=dropout)
        model.fit(x_train, y_train, 
                  epochs=10, batch_size=64, 
                  validation_data=(x_test, y_test), 
                  verbose=0, callbacks=[wandb.keras.WandbCallback()])

wandb.agent(sweep_id=sweep_id, function=train)
wandb.finish()
```

## Аналіз результатів

Із отриманих результатів, видно, насправді, не дуже багато. Проглядається, що Dropout шар трохи допоміг кривій втрат стати ще більш спадною і досягти ще меншого значення, але на метриках, тобто якості передбачень це не відобразилося.

У мене є підозра, що це мої помилки у ініціалізації методів регуляризації, а саме: неправильні місця розстановки Dropout (тільки після повнозв'язних у кінці), занадто строга регуляризації через L2 (на кожному шарі). Тому як майбутнє покращення, можна спробувати змінити акцент на ці аспекти.

Також, причиною може бути архітектура. Хоч і згорткових блоків багато, нейронів теж, як і пулінґових шарів, проте цього виявляється недостатньо, щоб ще більше покращити цю модель. Можливо, вона занадто просто для зображень такого розміру, ось тому і перенавчається так сильно та демонструє "стелю" у своїх метриках, попри покращення (зменшення) втрат як тренувальних, так і валідаційних.

Як останній варіант, можна спробувати здійснити transfer learning із якоюсь моделлю, наприклад, почати із чогось найбільш популярного (VGG16) та поглянути як це вплине на наші результати.

Ще один цікавий момент якого я не торкнувся — обрізання зображень та масштабування їх у залежності від їх пропорцій. Тому що як ми бачили у розділі Попередня обробка, зображення ми масштабували на квадратний формат 224х224, проте я знехтував тим, що у них присутні чорні рамки чи вони були розтягнуті вздовж вертикалі. Тому це могло зіграти, можливо, не ключову, але вагому роль у тренуванні наших моделей. Це неодмінно можна занести у перелік потенційних покращень поряд із тими про які ми говорили раніше.

Крім цього, варто звернути увагу на більш інтенсивну аугментацію, а не тільки для урівняння класів.

Також, не завадить розширити набір даних за допомогою додаткової аугментації обидвох класів та обстежити всі наявні початкові зображення для виявлення аномалій, які можна також грамотно обробити.
